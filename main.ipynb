```python
q = '''
SELECT specialization, "area.name" as city, avg_salary
FROM 
(SELECT specialization, "area.name", AVG(CAST(salary AS FLOAT)) as avg_salary, ROW_NUMBER() over (PARTITION BY specialization order by avg_salary desc) as rank
FROM df_cl
GROUP BY specialization, "area.name"
ORDER BY avg_salary DESC)
where rank <=3

'''
s = duckdb.query(q).to_df()
s
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.util import ngrams
import re
nltk.download('stopwords')
nltk.download('punkt')
# Загрузка стоп-слов
stop_words = set(stopwords.words('russian'))

# Функция для удаления стоп-слов и создания биграмм
def process_text(text):
    text = re.sub(r'[^\w\s]','', text)  # Удаление пунктуации
    tokens = word_tokenize(text.lower())  # Токенизация и приведение к нижнему регистру
    tokens = [word for word in tokens if word not in stop_words]  # Удаление стоп-слов
    bigrams = list(ngrams(tokens, 2))  # Создание биграмм
    return tokens, bigrams

# Применение функции к столбцу 'description.vacancy'
df_cl['tokens'], df_cl['bigrams'] = zip(*df_cl['description.vacancy'].apply(process_text))

# Вычисление частотности биграмм
count_vectorizer = CountVectorizer(analyzer='word', ngram_range=(2, 2))
X = count_vectorizer.fit_transform(df_cl['bigrams'].apply(lambda x: ' '.join(map(str, x))))
bigram_frequencies = X.toarray()

# Вычисление TF-IDF для каждой подгруппы
tfidf_vectorizer = TfidfVectorizer(analyzer='word', ngram_range=(2, 2))
tfidf_values = {}
for group, bigrams in df_cl.groupby('specialization')['bigrams']:
    tfidf_matrix = tfidf_vectorizer.fit_transform(bigrams.apply(lambda x: ' '.join(map(str, x))))
    tfidf_values[group] = tfidf_matrix

# Вывод результатов
for group, tfidf_matrix in tfidf_values.items():
    print(f"TF-IDF для группы {group}:\n{tfidf_matrix.toarray()}")
tfidf_values
```
